```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
''' Line Plot '''

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x,y)
plt.title("Sample Test")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
''' Bar Plot'''

categories = ['A', 'B', 'C']
values = [5, 7, 3]

plt.bar(categories, values)
plt.show()
```

```python
'''Scatter Plot'''
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='green')
plt.title("Scatter Plot")
plt.xlabel("Feature X")
plt.ylabel("Feature Y")
plt.show()
```

```python
''' Histogram '''
data = [1, 2, 2, 3, 3, 3, 4, 5, 5, 6]
plt.hist(data, bins=5, color='orange')
plt.title("Histogram")
plt.show()
```

```python
''' Subplots '''

plt.figure(figsize=(10, 7))

plt.subplot(2, 2, 1)
plt.plot([1, 2, 3], [4, 5, 6])
plt.title("Left Plot")

plt.subplot(2, 2, 2)
plt.bar(['A', 'B'], [5, 3])
plt.title("Right Plot")

plt.subplot(2, 2, 3)
plt.bar(['A', 'B'], [5, 3])
plt.title("Right Plot")

plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(8, 5))
plt.plot(x, y, color='red', linestyle='--', marker='o', label='Line')
plt.title("Styled Line Plot", fontsize=14, color='navy')
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.legend()
plt.show()
```

```python
df = pd.DataFrame({
    'City': ['Mumbai', 'Delhi', 'Chennai', 'Mumbai', 'Delhi'],
    'Depression': [1, 0, 1, 0, 1]
})

df['City'].value_counts().plot(kind='bar', color='teal')
plt.title("Students per City")
plt.show()
```
